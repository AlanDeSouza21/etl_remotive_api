```python
import requests
from dotenv import load_dotenv
import os
```

```python
load_dotenv()  
TOKEN = os.getenv("TOKEN")  

resposta = requests.get(
    "https://data.fantastic.jobs/v1/active-jb",
    headers={
        "Authorization": f"Bearer {TOKEN}"
    },
    params={
      "time_frame": "24h",
      "limit": "100",
    },
    timeout=10
)

print(resposta.status_code)
print(resposta.text)
```

```python
import json

dados = resposta.json()  # converte a resposta para dict/list Python

caminho = "VagasDados_remotive_api.json"

# 1. Lê o que já existe (se o arquivo existir)
if os.path.exists(caminho):
    with open(caminho, "r", encoding="utf-8") as f:
        dados_existentes = json.load(f)
else:
    dados_existentes = []

print(f"Vagas já salvas: {len(dados_existentes)}")

# 2. Junta com os dados novos (dados = resposta.json() da nova requisição)
dados_combinados = dados_existentes + dados

# 3. Remove duplicatas pelo campo 'id' (importante para não repetir vagas)
vistos = set()
dados_unicos = []
for vaga in dados_combinados:
    if vaga['id'] not in vistos:
        vistos.add(vaga['id'])
        dados_unicos.append(vaga)

# 4. Salva tudo de novo
with open(caminho, "w", encoding="utf-8") as f:
    json.dump(dados_unicos, f, ensure_ascii=False, indent=2)

print(f"Total após atualização: {len(dados_unicos)}")
```

```python
def extrair_pais(vaga):
    # 'vaga' é um dicionário representando uma vaga de emprego (um item da lista 'dados')
    
    # Pega a lista 'locations' dentro da vaga. Se não existir, usa uma lista com um 
    # dicionário vazio como padrão ([{}]), evitando erro de KeyError.
    # Depois pega o primeiro item dessa lista ([0]) e, dentro dele, o campo 'address'.
    # Se 'address' não existir, usa um dicionário vazio {} como padrão.
    address = vaga.get('locations', [{}])[0].get('address', {})
    
    # Tenta extrair o país diretamente do campo 'addressCountry'
    pais = address.get('addressCountry')
    
    # Se 'addressCountry' vier vazio, None, ou string vazia (todos avaliam como "falsy"),
    # tenta usar 'addressLocality' (cidade) como alternativa/fallback
    if not pais:
        pais = address.get('addressLocality')
    
    # Retorna o país encontrado (ou a cidade, como plano B, ou None se nada for encontrado)
    return pais

# Aplica a função extrair_pais() em cada vaga da lista 'dados'
# e cria/sobrescreve o campo 'paises' dentro de cada vaga com o resultado
for vaga in dados:
    vaga['paises'] = extrair_pais(vaga)

# Imprime o valor do campo 'paises' do primeiro item da lista, 
# só para conferir se a extração funcionou corretamente
print(dados[0]['paises'])  # testa no primeiro item
```

```python
#df['locations'][0]

def extrair_localizacao(locations):
    if locations and len(locations) > 0:
        endereco = locations[0].get('address', {})
        return pd.Series({
            'cidade': endereco.get('addressLocality'),
            'regiao': endereco.get('addressRegion'),
            'pais': endereco.get('addressCountry'),
        })
    return pd.Series({'cidade': None, 'regiao': None, 'pais': None})

# Remove qualquer coluna duplicada com esses nomes
df = df.drop(columns=['cidade', 'regiao', 'pais'], errors='ignore')

# Agora atribui de novo, limpo
df[['cidade', 'regiao', 'pais']] = df['locations'].apply(extrair_localizacao)

```
