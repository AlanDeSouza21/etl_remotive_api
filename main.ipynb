```python
import requests
from dotenv import load_dotenv
import os
```

```python
load_dotenv(override=True)  # <-- override=True força sobrescrever variáveis já carregadas
TOKEN = os.getenv("TOKEN")
print(f"Token carregado: {TOKEN[:10]}...")  # confirme que é o token novo
resposta = requests.get(
    "https://data.fantastic.jobs/v1/active-jb",
    headers={
        "Authorization": f"Bearer {TOKEN}"
    },
    params={
      "time_frame": "24h",
      "limit": 10
    }
)

print(resposta.status_code)
print(resposta.text)
```

```python
import json

dados = resposta.json()  # converte a resposta para dict/list Python

caminho = "/opt/airflow/dags/VagasDados_remotive_api.json"

# Verifica se a requisição foi bem-sucedida antes de continuar
if resposta.status_code != 200:
    print(f"Erro na requisição: {resposta.status_code}")
    print(resposta.json().get("detail", "Sem detalhes"))
else:
    dados = resposta.json()  # só converte se status for 200

    # Garante que dados é uma lista
    if isinstance(dados, dict):
        dados = dados.get("jobs", dados.get("data", []))  # ajuste a chave conforme a API

    # 1. Lê o que já existe
    if os.path.exists(caminho):
        with open(caminho, "r", encoding="utf-8") as f:
            dados_existentes = json.load(f)
    else:
        dados_existentes = []

    print(f"Vagas já salvas: {len(dados_existentes)}")

    # 2. Junta e remove duplicatas
    dados_combinados = dados_existentes + dados

    vistos = set()
    # ... resto do código
dados_unicos = []
for vaga in dados_combinados:
    if vaga['id'] not in vistos:
        vistos.add(vaga['id'])
        dados_unicos.append(vaga)

# 4. Salva tudo de novo
with open(caminho, "w", encoding="utf-8") as f:
    json.dump(dados_unicos, f, ensure_ascii=False, indent=2)

print(f"Total após atualização: {len(dados_unicos)}")
```

```python
# tratamento das colunas 
df_col_selecionadas['ai_salary_value'] = df_col_selecionadas['ai_salary_value'].round(2)
df_col_selecionadas.loc[:, 'date_posted'] = df_col_selecionadas['date_posted'].str.split('T').str[0]
df_col_selecionadas.loc[:, 'ai_salary_min_value'] = df_col_selecionadas['ai_salary_min_value'].fillna(0).round(2)
df_col_selecionadas.loc[:, 'ai_salary_value'] = df_col_selecionadas['ai_salary_value'].fillna(0).round(2)
df_col_selecionadas.loc[:, 'ai_salary_max_value'] = df_col_selecionadas['ai_salary_max_value'].fillna(0).round(2)
df_col_selecionadas.loc[:, 'id'] = df_col_selecionadas['id'].astype('int64')
df_col_selecionadas = df_col_selecionadas.rename(columns={'locations': 'country'})

# EVITA ERRO DE PROGRAMAÇÃO CAUSADOS POR VALORES NaN AO INSERIR OS DADOS NA TABELA DO BANCO MYSQL 
df_col_selecionadas = df_col_selecionadas.where(
    pd.notnull(df_col_selecionadas),
    None
) 

df_col_selecionadas.head(3)
#df_col_selecionadas.info()
```

```python
from sqlalchemy import create_engine, text
from sqlalchemy.engine import URL

load_dotenv()  
DB_HOST = os.getenv("DB_HOST")  
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_NAME = os.getenv("DB_NAME")
DB_CHARSET = os.getenv("DB_CHARSET")

engine = create_engine(URL.create(
    drivername="mysql+pymysql",
    username=DB_USER,
    password=DB_PASSWORD,
    host=DB_HOST,
    database=DB_NAME,
    query={"charset": DB_CHARSET}
))
```

```python
with engine.begin() as conn:
    conn.execute(text("""
        CREATE TABLE IF NOT EXISTS remotive_tabela_empregos (
                      id BIGINT PRIMARY KEY,
                      date_posted DATE,
                      title VARCHAR(255),
                      organization VARCHAR(255),
                      country VARCHAR(255),
                      ai_employment_type VARCHAR(255),
                      source_type VARCHAR(255),
                      seniority VARCHAR(255),
                      ai_salary_currency VARCHAR(255),
                      ai_salary_value DECIMAL(15,2),
                      ai_salary_min_value DECIMAL(15,2),
                      ai_salary_max_value DECIMAL(15,2),
                      ai_salary_unit_text VARCHAR(255),
                      ai_key_skills VARCHAR(255),
                      ai_taxonomies_a VARCHAR(255),
                      ai_education VARCHAR(255)
        )
    """))
```

```python
import numpy as np

def limpar_valor(v):
    if isinstance(v, float) and np.isnan(v):
        return None
    if isinstance(v, (np.int64, np.int32)):
        return int(v)
    if isinstance(v, (np.float64, np.float32)):
        return float(v)
    return v

with engine.begin() as conn:
    for i in range(len(df_col_selecionadas)):
        row = {col: limpar_valor(df_col_selecionadas[col][i]) for col in df_col_selecionadas.columns}
        conn.execute(text("""
            INSERT IGNORE INTO remotive_tabela_empregos (
                id, date_posted, title, organization, country,
                ai_employment_type, source_type, seniority,
                ai_salary_currency, ai_salary_value,
                ai_salary_min_value, ai_salary_max_value,
                ai_salary_unit_text, ai_key_skills,
                ai_taxonomies_a, ai_education
            ) VALUES (
                :id, :date_posted, :title, :organization, :country,
                :ai_employment_type, :source_type, :seniority,
                :ai_salary_currency, :ai_salary_value,
                :ai_salary_min_value, :ai_salary_max_value,
                :ai_salary_unit_text, :ai_key_skills,
                :ai_taxonomies_a, :ai_education
            )
        """), row)
```
